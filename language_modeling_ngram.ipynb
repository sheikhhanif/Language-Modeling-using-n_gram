```python
import re
from collections import Counter
import numpy as np


def ngram_train():

    #using "with" will close the file immediately after last call
    with open("data/text-ngram.dat","r") as rfile:
        infile = rfile.readlines()

    unifile = open("data/unigram.dat","w")
    bifile = open("data/bigram.dat","w")
    trifile = open("data/trigram.dat","w")   
    
    sent = [s for s in infile[0].split(".")]
   
    for line in sent:        
        for j in range(len(line.split())-1):
            uni = line.split()[j]
            bi = line.split()[j]+" "+line.split()[j+1]
            count = Counter(uni)
           
            
            unifile = open("data/unigram.dat","a")
            bifile = open("data/bigram.dat","a")            
            unifile.write(str(uni)+"\n")
            bifile.write(str(bi)+"\n")            
            unifile.close()
            bifile.close()

        for k in range(len(line.split())-2):
            trifile = open("data/trigram.dat","a")            
            tri = line.split()[k]+" "+line.split()[k+1]+" "+line.split()[k+2]
            trifile.write(str(tri)+"\n")
            trifile.close()
    
    # Creating unigram dictionary using counter   
    #print("Unigram dictionary with counter\n**********************************")
    with open('data/unigram.dat','r') as f:
        u = f.read()
    uni_count = dict(Counter(u.split("\n")))
    #print(uni_count)
    #print("\n\n")
    
    
    # Creating bigram dictionary using counter 
    #print("Bigram dictionary with counter\n*****************************")
    with open('data/bigram.dat','r') as f:
        bi = f.read()
    bi_count = dict(Counter(bi.split("\n")))
    #print(bi_count)
    #print("\n\n")
    
    
    # Creating trigram dictionary using counter 
    #print("Trigram dictionary with counter\n**********************************")
    with open('data/trigram.dat','r') as f:
        tri = f.read()
    tri_count = dict(Counter(tri.split("\n")))
    #print(tri_count)
    
    return (uni_count, bi_count, tri_count)






def ngram_test():
    
    """
    This function test bigram and trigram model in train ngram
    """
    
    # test sentece
    sentence = "KICT also aspires to enhance the quality of learning and teaching"
    
    # getting the dictionary of train uni, bi and tri gram
    uc, bc, tc = ngram_train()
    
    # getting token from the sentence
    tokens = sentence.split()
    
    # unigram of sentence
    uni_gram = Counter(tokens)

    # bigram
    bi_gram = [(tokens[i] + " " + tokens[i+1]) for i in range(0, len(tokens) -1)]
    
    # trigram
    tri_gram = [(tokens[i] + " " + tokens[i+1] + " " + tokens[i+2]) for i in range(0, len(tokens) -2)]

    # dictionary of unigram
    dic_ug = dict(Counter(uni_gram))
    
    # dictionary of bigram
    dic_bg = dict(Counter(bi_gram))
  
    # dictionary of trigram
    dic_tg = dict(Counter(tri_gram))
    
    # probility list for bigram model
    prob_bg = []
    for key in dic_bg:
        prob_bg.append(bc[key] / uc[key.split(" ")[0]])
    # total probability of bigram
    total_prob_bg = np.prod(np.array(prob_bg))
    
    # ent list for holding log value of each probability
    ent = []
    ent.append(np.log2(prob_bg))
    bi_entropy = 0
    bi_entropy = -1 * (np.sum(np.array(prob_bg) * np.array(ent)))
    
    print("Total probability of bigram model: ", total_prob_bg)
    print("Entropy for Bigram model: ", bi_entropy)
    print("Perplexity of bigram model: ", 2**bi_entropy)
    
    # Because we don't have zero probability in bigram model we don't apply smoothing
    
    # probability list for tri_gram
    # Using laplace smooting to deal with zero probability
    prob_tg = []
    for key in dic_tg:
        if key in tc:
            prob_tg.append((tc[key] + 1 )/ (bc[key.split(" ")[0] + " " + key.split(" ")[1]] + len(dic_ug)))
        if key not in tc:
            prob_tg.append(1 / (bc[key.split(" ")[0] + " " + key.split(" ")[1]] + len(dic_ug)))
    total_prob_tg = np.prod(np.array(prob_tg))
    
    print("\nProbability of trigram model after smoothing", total_prob_tg)
    
    
# calling ngram_test to show result
ngram_test()
```
